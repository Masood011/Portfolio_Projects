```python
df8 = df[['name','quote.USD.price']]
df8
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a clean background style
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(10, 6))
barplot = sns.barplot(
    x='name', 
    y='quote.USD.price', 
    hue='name',  # Assign the x variable to hue
    data=df8, 
    palette='viridis', 
    dodge=False  # Ensure bars aren't split by hue
)

if barplot.legend_ is not None:
    barplot.legend_.remove()

# Add titles and labels
plt.title('Bitcoin Prices', fontsize=16, fontweight='bold')
plt.xlabel('Cryptocurrency Names', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)

# Rotate x-axis labels for clarity
plt.xticks(rotation=45, fontsize=10, ha='right')

# Annotate bars with their values
for bar in barplot.patches:
    bar_height = bar.get_height()
    barplot.annotate(f'{bar_height:.2f}',  # Format the value to 2 decimal places
                     (bar.get_x() + bar.get_width() / 2, bar_height),
                     ha='center', va='bottom', fontsize=10, color='black')

# Show the plot
plt.tight_layout()
plt.show()
```
